```python
import pandas as pd
```

# GBM

```python
df_ko = pd.read_csv('./GBM/Pseudo_GBM_KO_to_UniRef90.csv', sep = '\t', header=None)
df_ko.head()
```

```python
df_tigrfam = pd.read_csv('./GBM/Pseudo_GBM_TIGRFAM_to_UniRef90.csv', sep = '\t', header=None)
df_tigrfam.head()
```

```python
df_eggnog = pd.read_csv('./GBM/Pseudo_GBM_EGGNOG_to_UniRef90.csv', sep = '\t', header=None)
df_eggnog.head()
```

```python
Pesudo_map = pd.concat([df_ko, df_tigrfam, df_eggnog])
```

# Bacterial annotation of the 2 Missing reactions of GMM can not be found on Uniprot, so are TIGR01269 and TIGR03812 in GBM, we therefore exclude them from the pesudo mapping file

```python
Pesudo_map = Pesudo_map[~Pesudo_map[0].str.contains('TIGR01269|TIGR03812')]
```

```python
Pesudo_map.to_csv('./Pesudo_map_eggnong_ko_tigrfam_uniref90.txt', sep='\t',index=False, header=False)
```
